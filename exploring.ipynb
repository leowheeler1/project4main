```python
# Data Processing
import pandas as pd
import numpy as np

# Modelling
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, precision_score, recall_score, ConfusionMatrixDisplay
from sklearn.model_selection import RandomizedSearchCV, train_test_split
from scipy.stats import randint
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LinearRegression
```

```python
df1 = pd.read_csv('datasets/sdTest.csv')
df2 = pd.read_csv('datasets/sdTrain.csv')
df3 = pd.concat([df1, df2])
df3
```

```python
symptoms = []
for i in range(len(df3['text'])):
    string = df3.iloc[i,0].split(',')
    for symptom in string:
        if ' ' not in symptom:
            if '"' not in symptom:
                if symptom not in symptoms:
                    symptoms.append(symptom)
symptoms
```

```python
for symptom in symptoms:
    df3[symptom] = [0 for i in range(len(df3['text']))]
df3
```

```python
columns = list(df3.columns)
col_pos = {}
for i in range(len(columns)):
    col_pos[columns[i]] = i
```

```python
for i in range(len(df3['text'])):
    symptom_list = df3.iloc[i,0].split(',')
    for column in df3.columns:
        if column in symptom_list:
            df3.iloc[i,col_pos[column]] = 1
```

```python
df3
```

```python
df3.drop('text', axis=1, inplace=True)
```

```python
df3.rename(columns={'label':'score'}, inplace=True)
df3
```

```python
# bin the score
def binner(scores):
    output = []
    for score in scores:
        if score > 900:
            output.append(1)
        elif score > 650:
            output.append(2)
        elif score > 400:
            output.append(3)
        elif score > 200:
            output.append(4)
        else:
            output.append(5)
    return output
```

```python
df3['target'] = binner(df3['score'])
df3
```

```python
df3.drop('score', axis=1, inplace=True)
```

```python
df3.to_csv('datasets/OneHotEncData.csv', index=False)
```

```python
df3
```

```python
df4 = df3.copy()
y = df4['target']
df4.drop('target', axis=1, inplace=True)
X = df4
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
model = RandomForestClassifier()
model.fit(X_train, y_train)
```

```python
preds = model.predict(X_test)
preds
```

```python
accuracy_score(y_test, preds)
```

```python
precision_score(y_test, preds, average='weighted')
```

```python
recall_score(y_test, preds, average='weighted')
```
