```python
# importing dependencies
import sys
import os
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report
```

```python
# adding directory and path
sys.path.append(os.path.abspath("Resources"))
```

```python
# loading data and vectorizer
from transformdata import X_train, X_test, y_train, y_test, vectorizer
print(X_train)
```

```python
# training NB model
model = MultinomialNB()
model.fit(X_train, y_train)
```

```python
# evaluating accuracy
y_pred = model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
# defining chatbot feedback
def classify_input(user_input):
    input_vec = vectorizer.transform([user_input])
    prediction = model.predict(input_vec)[0]

    if prediction == 1:
        return "ðŸš¨ URGENT! ðŸš¨ Please seek immediate help!"
    elif prediction == 2:
        return "Somewhat urgent. Please see a medical professional as soon as you are able."
    elif prediction == 3:
        return "Moderate. Please closely monitor your symptoms."
    elif prediction == 4:
        return "Mild. Please ensure that you are well-rested and keep track of your symptoms."
    else:
        return "Non-urgent. You seem to be in a good state, but stay alert for any further symptoms."
```

```python
# catching chatbot errors
if __name__ == "__main__":
    while True:
        user_input = input("\nTell me your symptoms (or type 'exit' to end your session): ")
        if user_input.lower() == "exit":
            break
        try:
            result = classify_input(user_input)
            print(result)
        except Exception as e:
            print("Error", e)
```
